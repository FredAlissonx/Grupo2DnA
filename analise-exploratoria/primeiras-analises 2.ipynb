```python
import pandas as pd
```

```python
# Importando Base de Dados
df = pd.read_csv('C:/Users/CB517RL/OneDrive - EY/Desktop/Projeto Academy/churn.csv')
df.head()
```

```python
# Excluindo colunas de sobrenome e index 
df = df.drop(columns=["Surname", "RowNumber"])
df.head()
```

```python
# Categorizando as variaveis
df['id_geography'] = df['Geography'].replace({'France': 0, 'Spain': 1, 'Germany': 2})
df['id_gender'] = df['Gender'].replace({'Female': 0, 'Male': 1})
df.head()
```

```python
import matplotlib.pyplot as plt

# ... Assumindo que df é o seu DataFrame

# Crie uma nova coluna 'ageRange' com a age dividida por 10, multiplicada por 10 e convertida para um inteiro
df['ageRange'] = (df['Age'] // 10 * 10).astype(int)

# Agora, crie um gráfico de barras da frequência de cada range de idade
age_frequency = df['ageRange'].value_counts().sort_index()

age_frequency.plot(kind='bar', figsize=(10,6))
plt.xlabel('Age Range')
plt.ylabel('Frequency')
plt.title('Frequency of Age Ranges')
plt.show()
```
